## Предобработка данных

```python
#Посмотрим на процентное соотношение всех пропущенных значениях по всему датасету
df.isnull().sum().sort_values()/df.shape[0]*100
     
```

```python
#удалим строки в столбцах 'locality_name'  и 'floors_total', где отсутствуют значения, т.к. они  не критичны
#методом dropna()
df.dropna(subset=['locality_name'], inplace=True)
df.dropna(subset=['floors_total'], inplace=True)
```

```python
#выведем названия населенных пунктов
df['locality_name'].unique()
```

```python
#заменим ё на е в названиях 
df['locality_name'] = df['locality_name'].str.replace("ё", "е") 
#уберем тип поселения, оставим только названия
df["locality_name"] = df["locality_name"].str.strip().str.replace("деревня|поселок|село|городской|городского типа|садовое товарищество|коттеджный|при железнодорожной станции|садоводческое некоммерческое товарищество", "", regex=True)
df["locality_name"] = df["locality_name"].replace(r'\s+','',regex=True)
df['locality_name'].unique()
```

```python
#Заполнение жилой и кухонной зоны медианой может привести к тому, что сумма жилой и кухонной площадей будет больше,
#чем общая площадь наблюдения, что вообще нереально
#Поэтому заменим пустые значения соотношением  средних величин жилой и общей площади и соотношением ср.кухни и общей площади
living_ratio = df['living_area'].mean() / df['total_area'].mean()
kitchen_ratio = df['kitchen_area'].mean() / df['total_area'].mean()
df['living_area'].fillna(living_ratio * df['total_area'], inplace=True)
df['kitchen_area'].fillna(kitchen_ratio * df['total_area'], inplace=True)
```

```python
# Пропущенных значений  в столбце 'is_apartment' 88%. Значит, это не апартаменты, что логично
#Заменим пропущенные значения на False
df['is_apartment'] = df['is_apartment'].fillna(value=False)
```

```python
#т.к. медианное и среднее значение в потолках не сильно отличаются,
#а на среднее могли повлиять аномально высокие значения высот потолков,
#то заполним пропуски по медианному значению
median_height=df['ceiling_height'].median()
df['ceiling_height'].fillna(median_height,inplace=True)
round(df['ceiling_height'].describe(),2)
```

```python
# еще раз посмотрим на выбросы в высотах потолка (при высоте > 6 метров) по этажам
df_grouped = df.groupby('floor').agg({'ceiling_height':['count', 'mean', lambda x: sum(x>6)]})
df_grouped
```

```python
#Построим гистограмму, чтобы наглядно посмотреть на высокие значения потолков
df.plot(y = 'ceiling_height', kind = 'hist', bins = 30, range = (2,6), grid=True, figsize = (5,3))
plt.show()
```

Высокие и низкие значения высот потолка не влияют особо на выборку. Пропустим их.

```python
#Число балконов: пропущено 48,6% значений.
#Предположим, что отсутствие числа балконов свидетельствует о том, что число балконов равно нулю.
#заполним пропущенные значения с помощью метода 'fillna()'
df['balcony'].fillna(0, inplace=True)
df['balcony'].isna().value_counts()
```

```python
#Посчитаем и добавим цену квадратного метра
df['cost_sqm'] = round(df['last_price'] / df['total_area'], 1)
```

```python
#Добавим день недели, месяц и год публикации объявления
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
```

```python
#Добавим этаж квартиры: первый, последний, другой
#напишем функцию категоризации по этажам, используя метод apply применимо к каждой строке датафрейма через axis = 1
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    elif 1 < floor < floors_total:
        return 'другой'
    
#категоризуем этажи с помощью функции  
df['floor_category'] = df.apply(floor_category, axis = 1)
```

```python
#Посчитаем и добавим соотношение жилой и общей площади, а также отношение площади кухни к общей
df['living_area_ratio'] = round(df['living_area'] / df['total_area'] * 100, 3)
df['kitchen_area_ratio'] = round(df['kitchen_area'] / df['total_area'] * 100, 3)
```

```python
# Функция для подсчёта границ с учетом 1.5 размаха
def det_whiskers(column):
    stat = df[column].describe()
    # межквартирльный размах
    iqr = stat[6] - stat[4]
     # левый и правый "ус"    
    left_whisker = round(stat[4] - 1.5 * iqr, 2)
    right_whisker = round(stat[6] + 1.5 * iqr, 2)

    # левый "ус" не должен быть меньше минимального значения
    if left_whisker < stat[3]: left_whisker = stat[3] 
        
    # правый "ус" не должен быть больше максимального значения
    if right_whisker > stat[7]: right_whisker = stat[7]
        
    return [left_whisker, right_whisker]
```

```python
# зададим функцию, вычисляющую долю аномальных значений
def anomalies_ratio(column):
    return 'Аномальные значения составляют {:.1%}'. format(len(df.loc[df[column] > det_whiskers(column)[1]]) / len(df)) + ' от всей выборки'
```

```python
#Выведем значения для площади
print(det_whiskers('total_area'))
anomalies_ratio('total_area')
```

```python
#Выведем значения для цены
print(det_whiskers('last_price'))
anomalies_ratio('last_price')
```

```python
print(det_whiskers('days_exposition'))
anomalies_ratio('days_exposition')
```

```python
print(det_whiskers('ceiling_height'))
anomalies_ratio('ceiling_height')
```

```python
print(det_whiskers('rooms'))
anomalies_ratio('rooms')
```
